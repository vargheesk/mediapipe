```python
import cv2
```

```python
mp_face = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
```

```python
face_det = mp_face.FaceDetection()
```

```python
video = cv2.VideoCapture(0)
while True:
    Suc,img = video.read()
    img = cv2.flip(img,1)
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result = face_det.process(img1)
    print(result.detections)
    # if result:

    cv2.imshow("webcam",img)
    if cv2.waitKey(1) & 0xff ==ord("q"):
        break
video.release()
cv2.destroyAllWindows()
```

```python
mp_face_mesh = mp.solutions.face_mesh
mp_draw = mp.solutions.drawing_utils
face_mesh = mp_face_mesh.FaceMesh()
```

```python
video = cv2.VideoCapture(0)
while True:
    suc, img = video.read()
    img = cv2.flip(img, 1)

    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(img1)

    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            mp_draw.draw_landmarks(img,face_landmarks)

    cv2.imshow("Face Mesh", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
mesh_style = mp_draw.DrawingSpec(
    color=(180, 180, 180),  # soft light gray
    thickness=1,
    circle_radius=0         # IMPORTANT: no dots
)
```

```python
video = cv2.VideoCapture(0)
while True:
    suc, img = video.read()
    img = cv2.flip(img, 1)

    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(img1)

    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            mp_draw.draw_landmarks(img,
                                   face_landmarks,
                                   connections=mp_face_mesh.FACEMESH_TESSELATION,
                                   landmark_drawing_spec=None,
                                   connection_drawing_spec=mesh_style)

    cv2.imshow("Face Mesh", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
